```python
import numpy as np

#Question 1
arrayQ1 = np.full((2,2),10)
print(arrayQ1)
```

```python
#Question 2
arrayQ2 = np.random.random((3,4))
print(arrayQ2)
```

```python
#Question 3
arrayQ3one = np.ones((3,8))
arrayQ3zero = np.zeros((5,3))
print("Array of ones: \n",arrayQ3one)
print("\nArray of zeros: \n", arrayQ3zero)
```

```python
#Question 4
arrayQ4 = np.random.random((6,3))
print("Array before transpose: \n",arrayQ4)
print("\nArray after transpose: \n",arrayQ4.T)
```

```python
#Question 5
n = int(input("Enter n: "))
p = int(input("Enter p: "))
m = int(input("Enter m: "))
arrayQ5np = np.random.random((n,p))
arrayQ5pm = np.random.random((p,m))
print("\nFirst array: \n",arrayQ5np)
print("\nSecond array: \n",arrayQ5pm)
print("\nDot Product: \n",np.dot(arrayQ5np,arrayQ5pm))
```
